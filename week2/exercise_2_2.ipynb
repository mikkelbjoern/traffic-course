```python
# Load in the excel data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load in the data
data = pd.read_excel('Week2_Exercise2_data.xlsx')
data
```

```python
# Plot the distribution of the Trips1 variable

plt.hist(data['Trips1'], bins=20, color='blue', edgecolor='black', label='Trips1')
plt.xlabel('Trips1')

plt.show()


```

```python
# Do summary statistics of the Trips1 variable 
data['Trips1'].describe()
```

```python
# Plot correlation matrix
import seaborn as sns

modified_dataset = data.copy()
# Drop the original nun-censored Trips variable
modified_dataset.drop('Trips', axis=1, inplace=True)

corr = modified_dataset.corr()

# Ignore the diagonal values
for i in range(corr.shape[0]):
    corr.iloc[i,i] = np.nan

# Use blue as good fit, red as bad fit
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.show()
```
