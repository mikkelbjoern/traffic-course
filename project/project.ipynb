```python
import matplotlib.pyplot as plt
import pandas as pd 

# Read in the data
data = pd.read_excel('PFI_2025_ex1_data.xlsx')
# Data types:
# TransportType: {1,2,4,6} (1=walk, 2=bike, 4=car, 6=pulic transport)

# Variables in the data
# UniqueID: int
# Gender: {1, 2}
# Age: int
# Mode: TransportType
# Distance: float
# Alternative: TransportType
# TravelCost: float
# choice: TransportType

# TT{N} (eg. TT1): Travel time for mode N
# TC{N} (eg. TC1): Travel cost for mode N
data


import statsmodels.api as sm

# Add an intercept
data['intercept'] = 1

# Create dummy variables for the mode
data = pd.get_dummies(data, columns=['Mode'], drop_first=True) 
# By dropping the first column we avoid the dummy variable trap setting walking as the reference category
# We can thus think of Mode_2 as how much more biking is preffered to walking.

data
```

```python
# Implementing the first logic models based on the linear utility function
# V(n,i) = k_i + a * cost(n,i) + b * time(n,i)
# where n 

X1 = data[['intercept', 'TravelCost', 'TT2', 'TT4', 'TT6']]
y1 = data['choice']
model1 = sm.MNLogit(y1, sm.add_constant(X1)).fit()
print(model1.summary())
```
