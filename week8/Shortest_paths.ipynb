## Exercise 1 week 5: Shortest path tree ##

Consider the following network, where each link is directional and where the cost of traversing each link is as shown (e.g. 2 between 1 and 3).


![](Network.png)

Using *networkx* Calculate the shortest path from 1 to all other nodes, thereby constructing the shortest path tree from 1.

```python
## Import networkx 
import networkx as net
```

```python
## Create our graph using the text file input 'Example_network.txt'

FileName="Example_network.txt"
Graphtype=net.DiGraph()   # our graph is directed, we can use net.Graph() for an undirected graph

# For graph types, see
# https://networkx.org/documentation/stable/reference/classes/index.html#basic-graph-types

G = net.read_edgelist(
    FileName, 
    create_using=Graphtype,
    nodetype=int,
    data=(('weight',int),) # use float for real valued edge weights
)

## For reading and writing a graph see documentation here:
# https://networkx.org/documentation/stable/reference/readwrite/index.html
```

```python
# Print the list of nodes
G.nodes()
```

```python
# Print the list of edges
G.edges()
```

```python
# Print the edge weights
# Find the weight for each node
for u,v in G.edges():
      print ("Weight of Edge ("+str(u)+","+str(v)+")", G.get_edge_data(u,v))
```

```python
# Visualize the graph
import matplotlib as plt
pos = net.shell_layout(G)
net.draw(G,pos,with_labels=True)
```
