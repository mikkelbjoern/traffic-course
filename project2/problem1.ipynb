### Plots for 1.1.c

```python
import numpy as np
import matplotlib.pyplot as plt


v_0 = 70 # km/h
k_jam = 200 # veh/h

def v_greenshield(k):
    return v_0 * (1 - k/k_jam)

def q_greenshield(k):
    return k * v_greenshield(k)


# From 1.1.b 
k_max = 0.5 * k_jam
q_max = q_greenshield(k_max)

ks = np.linspace(0, k_jam, 200)
vs = v_greenshield(ks)
qs = q_greenshield(ks)

# Plot the (q, v), (v,k), (q,k) diagrams
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].plot(ks, qs, label='Greenshield fundamental diagram')
ax[0].set_xlabel(r'$k$ (veh/km)', fontsize=12)
ax[0].set_ylabel(r'$q$ (veh/h)', fontsize=12)
ax[0].set_title(r'$q(k)$', fontsize=14)
ax[0].grid()
# Limit to positive values
ax[0].set_xlim(0, k_jam * 1.05)
ax[0].set_ylim(0, q_max * 1.05)

ax[1].plot(qs, vs, label='Greenshield fundamental diagram')
ax[1].set_xlabel(r'$q$ (veh/h)', fontsize=12)
ax[1].set_ylabel(r'$v$ (km/h)', fontsize=12)
ax[1].set_title(r'$v(q)$', fontsize=14)
ax[1].grid()
# Limit to positive values
ax[1].set_xlim(0, q_max * 1.05)
ax[1].set_ylim(0, v_0 * 1.05)

ax[2].plot(ks, vs,  label='Greenshield fundamental diagram')
ax[2].set_xlabel(r'$k$ (veh/km)', fontsize=12)
ax[2].set_ylabel(r'$v$ (km/h)', fontsize=12)
ax[2].set_title(r'$v(k)$', fontsize=14)
ax[2].grid()
# Limit to positive values
ax[2].set_xlim(0, k_jam * 1.05)
ax[2].set_ylim(0, v_0 * 1.05)

# Add the critical point
ax[0].plot(k_max, q_max, 'ro', label='Critical point')
ax[1].plot(q_max, v_greenshield(k_max), 'ro', label='Critical point')
ax[2].plot(k_max, v_greenshield(k_max), 'ro', label='Critical point')

for a in ax:
    a.legend()

plt.tight_layout()
plt.show()
```

```python
# Confirming that derived formula is correct
print("Calculated critical point:", k_max, q_max)
q_max_derived = k_max * v_0 - v_0 * k_max**2 / k_jam
print("Derived critical point:", k_max, q_max_derived)
```
